```python
import numpy as np
import pandas as pd
```

```python
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import cross_val_score,KFold
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor()
```

```python
kfold = KFold(n_splits=5, shuffle=True, random_state=1)
scores = cross_val_score(knn, X, y, cv=kfold, scoring='r2')
```

```python
scores.mean()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
y, test_size=0.33, random_state=42)
```

### GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
knn = KNeighborsRegressor()
```

```python
param_grid = {
    'n_neighbors':[1,3,5,7,10,12,15,17,20],
    'weights':['uniform','distance'],
    'algorithm':['ball_tree', 'kd_tree', 'brute'],
    'p':[1,2]
}
```

```python
gcv = GridSearchCV(estimator=knn, param_grid=param_grid, scoring='r2', refit=True, cv=kfold, verbose=2)
```

```python
gcv.fit(X_train,y_train)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

```python
gcv.cv_results_
```

```python
pd.DataFrame(gcv.cv_results_)[['param_algorithm',	'param_n_neighbors',	'param_p', 'param_weights', 'mean_test_score']].sort_values('mean_test_score',ascending=False)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred = gcv.predict(X_test)
print(r2_score(y_test , y_pred))
```

### RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rcv = RandomizedSearchCV(estimator=knn, param_distributions=param_grid, scoring='r2', refit=True, cv=kfold, verbose=2)
```

```python
rcv.fit(X_train,y_train)
```

```python
rcv.best_score_
```

```python
rcv.best_params_
```

```python
y_pred = rcv.predict(X_test)
print(r2_score(y_test , y_pred))
```
