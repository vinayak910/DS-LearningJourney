```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate the data
x = 10 * np.random.rand(50) # -> input

y = 3 * x - 8 + np.random.randn(50) * 4 # -> # True equation of population f() + E

# where f() is 3*x - 8   (B1*x + B0)   , here B1 is 3 and B0 = -8.
# here E (irreducible error) is np.random.randn(50) * 4
```

now the function we will find is f'(x) , in this we try to find the closest to true values of coef ie B1 = 3 and B0 = -8
- ie try to make our function most similar to the true function(population)

```python
x = x.reshape(-1, 1)
model = LinearRegression()
model.fit(x, y)
```

```python
# Calculate the predicted values
y_pred = model.predict(x)
```

plotting the line for true function(population) ie for B0 and B1 vs The function we made from the data we have ie b0 and b1

```python
# Plot the scatter plot and regression lines
plt.scatter(x, y, label="Data points")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Scatter plot with increased variability and regression lines")

# Plot the actual population line
x_line = np.linspace(0, 10, 100)
y_actual = 2 * x_line - 5
plt.plot(x_line, y_actual, 'r', label="Population line (m=3, b=-8)")

# Plot the estimated regression line
y_estimated = model.coef_[0] * x_line + model.intercept_
plt.plot(x_line, y_estimated, 'g', label=f"Estimated line (m={model.coef_[0]:.2f}, b={model.intercept_:.2f})")

# Add legend and show the plot
plt.legend()
plt.show()
```

The true population line is red with coef of 3 and -8 but due to the data we have which also have randomness , due to this our estimation is different from population

- that's where statistic comes in basically you will do hypothesis testing and other things , which will tell if the m and b from sample we have is useful or not
-  we can calculate confidence interval , that this is the range in which the m and b lies
