# Goodness of Fit  (manually)

Ques1 . Suppose we have a six-sided fair die, and we want to test if the die is indeed fair. We roll the 
die 60 times and record the number of times each side comes up. We'll use the Chi-Square 
Goodness-of-Fit test to determine if the observed frequencies are consistent with a fair die 
(i.e., a uniform distribution of the sides).
Observed frequencies:
- Side 1: 12 times
- Side 2: 8 times
- Side 3: 11 times
- Side 4: 9 times
- Side 5: 10 times
- Side 6: 10 times

H0 : Dice follows a uniform distribution 	 , H1 : Dice doesnt follow uniform distribution

```python
import scipy.stats as stats

test_statistic = 1  # Chi-Square test statistic 
degrees_of_freedom = 5  # Degrees of freedom 

# Calculate the p-value using the chi2 survival function (sf)
p_value = stats.chi2.sf(test_statistic, degrees_of_freedom)

print("P-value:", p_value)
```

**Since P- value is more than alpha(0.05) , we cannot reject the null hypothesis . Hence H0 is true , data follows a uniform distribution**

# Goodness of fit (using stats function)

Ques 2 : Goodness of fit test for P class 

H0: P class follows uniform distribution
H1 : P class doesnt follow uniform distribution 

```python
import pandas as pd
import numpy as np
from scipy.stats import chisquare

# Load the Titanic dataset
data = pd.read_csv('train.csv')

data.head()
```

```python
# Count passengers in each class
class_counts = data['Pclass'].value_counts().sort_index()
class_counts

```

```python
# Calculate the expected counts assuming a uniform distribution
total_passengers = len(data)
expected_counts = total_passengers / 3
expected = [expected_counts] * 3
print("Expected Passenger Counts (assuming uniform distribution):\n", expected)

```

```python
# Perform the Chi-Square Goodness of Fit test
chi2, p_value = chisquare(class_counts, expected)

# Print the results
print("\nChi-Square Statistic: {:.2f}".format(chi2))
print("P-value: {:.4f}".format(p_value))

```

```python
# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("\nWe reject the null hypothesis. The distribution of passengers among the classes is not uniform.")
else:
    print("\nWe fail to reject the null hypothesis. The distribution of passengers among the classes is uniform.")
```

# Test for Independence 

problem statement -  We will use the Chi-Square test for independence to see if the survival rate of passengers is independent of the passenger class.

Null hypothesis= H0: Pclass and Survival are independent 
Alternative Hypothesis = H1 : Pclass and Survival arent independent

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Load the Titanic dataset
data = pd.read_csv('train.csv')

data.head()
```

```python
contingency_table = pd.crosstab(data['Survived'], data['Pclass'])
contingency_table
```

```python
# Perform the Chi-Square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("\nChi-Square Statistic: {:.2f}".format(chi2))
print("P-value: {:.4f}".format(p_value))
print("Degrees of Freedom: {}".format(dof))
print("Expected Frequencies: \n{}".format(expected))

```

```python
# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("\nWe reject the null hypothesis. There is a significant association between passenger class and survival rate.")
else:
    print("\nWe fail to reject the null hypothesis. There is no significant association between passenger class and survival rate.")
```
