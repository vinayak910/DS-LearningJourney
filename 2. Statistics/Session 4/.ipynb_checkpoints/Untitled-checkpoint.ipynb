# Standard Normal Variate

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
sns.kdeplot(df['Age'])
```

- This distribution is similar to normal distribution
- now time to convert into standard normal variate
- since the distribution is not exactly normal distribution 
- therefore , after standardizing the mean will not be exactly zero , but still close to zero

```python
# STANDARDIZING 
x = (df['Age']- df['Age'].mean())/df['Age'].std()
```

```python
x.mean() # close to zero
```

```python
x.std() # close to one
```

```python
sns.kdeplot(x)
```
