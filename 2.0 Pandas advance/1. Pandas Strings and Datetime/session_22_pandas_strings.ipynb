```python
# import titanic
df = pd.read_csv('train.csv')
df['Name']
```

```python
# Common Functions
# lower/upper/capitalize/title
df['Name'].str.upper()
df['Name'].str.capitalize()
df['Name'].str.title()
# len
df['Name'][df['Name'].str.len() == 82].values[0]
# strip
"                   nitish                              ".strip()
df['Name'].str.strip()
```

```python
# split -> get
df['lastname'] = df['Name'].str.split(',').str.get(0)
df.head()
```

```python
# another way using lambda instead of str accessor
print(df['Name'].apply(lambda x: x.split(',')[0])[0:5]) #---> last name
print(df['Name'].apply(lambda x: x.split(',')[1].split('.')[1])[0:5])#---> last name
print(df['Name'].apply(lambda x: x.split(',')[1].split('.')[0])[0:5])#---> title
```

```python
df['Name'].str.split(',').str.get(1).str.split('.', expand = True)
# now you can store both of them in new columns at once
```

```python
df[['title','firstname']] = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ', n=1, expand=True)
df.head()

df['title'].value_counts()
```

```python
# replace
df['title'] = df['title'].str.replace('Ms.','Miss.')
df['title'] = df['title'].str.replace('Mlle.','Miss.')
```

```python
df['title'].value_counts()
```

```python
# filtering
# startswith/endswith
df[df['firstname'].str.endswith('A')]
```

```python
# isdigit/isalpha...
df[df['firstname'].str.isalnum()]
```

```python
# applying regex
# contains
# search john -> both case
df[df['firstname'].str.contains('john',case=False)]
# find lastnames with start and end char vowel
df[df['lastname'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]
```
